```python
NEGATION_WORDS = {"not", "no", "never", "n't", "neither", "nor", "nobody", "nothing", "nowhere"}
PUNCTUATION = {'.', ',', '!', '?', ';', ':', '--'}
NEGATION_WINDOW = 3  # only tag the next N tokens after a negator

def tag_negations(sentence: str) -> str:
    tokens = sentence.split()
    result = []
    negation_countdown = 0

    for token in tokens:
        if token in NEGATION_WORDS:
            negation_countdown = NEGATION_WINDOW
            result.append(token)
        elif any(p in token for p in PUNCTUATION):
            negation_countdown = 0
            result.append(token)
        elif negation_countdown > 0:
            result.append(f"NOT_{token}")
            negation_countdown -= 1
        else:
            result.append(token)

    return " ".join(result)
```

```python
def lemmatize(sentence: str) -> str:
    if not isinstance(sentence, str):
        return ""

    def get_wordnet_pos(treebank_tag):
        from nltk.corpus import wordnet
        if treebank_tag.startswith('J'):
            return wordnet.ADJ
        elif treebank_tag.startswith('V'):
            return wordnet.VERB
        elif treebank_tag.startswith('N'):
            return wordnet.NOUN
        elif treebank_tag.startswith('R'):
            return wordnet.ADV
        else:
            return None

    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)

    lemmatized = [
        wordnet_lemmatizer.lemmatize(word, get_wordnet_pos(tag)) if get_wordnet_pos(tag) else word
        for word, tag in pos_tags
    ]

    res = ' '.join(lemmatized)
    return res
```

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower() # smaller vocab space, probably fine because ALL CAPS is merely an amplifier and doesn't flip classification
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove special characters and punctuation for smaller vocab space. But random special characters !@&^#^&$#^$ might imply rage/negative sentiment
    text = re.sub(r'\d+', '', text) # Remove numbers
    tokens = text.split()
    # tokens = [word for word in tokens if word not in stopwords] # Remove specially curated stopword list
    return ' '.join(tokens)
```

```python
# Naive Bayes pipeline

MAX_FEATURES = 10000
ALPHA = 1.0
NGRAM_RANGE = (1, 3)
MIN_DF = 2

def clean(series: pd.Series) -> pd.Series:
    return (series
        # .apply(lemmatize)
        .apply(preprocess_text)
        .apply(tag_negations)
    )

tfidf_vectorizer = TfidfVectorizer(max_features=MAX_FEATURES, ngram_range=NGRAM_RANGE, min_df=MIN_DF)


X_train_cleaned = clean(X_train)
X_train_vectorized = tfidf_vectorizer.fit_transform(X_train_cleaned)

X_test_cleaned = clean(X_test)
X_test_vectorized = tfidf_vectorizer.transform(X_test_cleaned)

model = MultinomialNB(alpha=ALPHA)
model.fit(X_train_vectorized, y_train)

print("Naive Bayes model trained successfully.")
```

```python
# "Naive" Naive Bayes
class NaiveBayesClassifier:
    priors = [0, 0, 0]
    likelihoods = [{}, {}, {}]
    vocab = set()

    def fit(self, X, y):

        # Words by category with duplicates (for frequency calculations)
        words_by_category = [[],[],[]]

        # All words with duplicates (for frequency calculations)
        all_words = ["UNK"]

        # Get counts
        category_counts = [1, 1, 1]
        prior_denom = len(X) + 3

        for i in range(len(X)):
            words1 = X[i].split(" ")
            words1 = [word.lower() for word in words1]
            words = [word for word in words1 if word.lower() not in stopwords]

            # Increment the correct category
            label = y[i]
            words_by_category[label].extend(words)
            category_counts[label] += 1

            # Increment the list of words found
            all_words.extend(words)

        # Get priors
        for i in range(3):
          self.priors[i] = category_counts[i]/prior_denom

        print(self.priors)

        # Get vocab
        self.vocab = set(all_words)

        # Frequency list (all words and per category)
        all_words = Counter(all_words)

        word_freq = [0, 0, 0]
        for i in range(3):
          word_freq[i] = Counter(words_by_category[i])

        # Calculate Likelihoods
        for word in all_words:
            likelihood_denom = all_words[word] + len(all_words)

            # For each category, find frequency of word in said category
            for i in range(3):
              likelihood = word_freq[i][word] + 1
              self.likelihoods[i][word] = likelihood/likelihood_denom

        print(self.likelihoods)

    def predict(self, X):
        y = []
        for i in range(len(X)):
            words1 = X[i].split(" ")
            words1 = [word.lower() for word in words1]
            words = [word for word in words1 if word.lower() not in stopwords]

            log_probs = [0, 0, 0]
            # Log Probability (Sum logs = Product probability)
            for i in range(3):
              log_probs[i] = math.log(self.priors[i])

            for word in words:
                if word not in self.vocab:
                    word = "UNK"

                for i in range(3):
                  log_probs[i] += math.log(self.likelihoods[i][word])

            y.append(log_probs.index(max(log_probs)))
        return y
```

```python
# Roberta pipeline

MAX_LEN = 64 # max length before the text is clipped
BATCH_SIZE = 32
EPOCHS = 1
LR = 2e-5
WEIGHT_DECAY = 0.01
# MODEL = "roberta-base" # large base model
MODEL = "distilroberta-base" # smaller model for iteration

class SentimentDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = np.array(texts)
        self.labels = np.array(labels)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            str(self.texts[idx]),
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'label': torch.tensor(self.labels[idx], dtype=torch.long)
        }

tokenizer = RobertaTokenizer.from_pretrained(MODEL)
train_dataset = SentimentDataset(X_train, y_train, tokenizer, MAX_LEN)
val_dataset = SentimentDataset(X_test, y_test, tokenizer, MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)


model = RobertaForSequenceClassification.from_pretrained(
    MODEL,
    num_labels=num_labels
)
model.to(device)
optimizer = AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
total_steps = len(train_loader) * EPOCHS
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.1 * total_steps),
    num_training_steps=total_steps
)

def train_epoch(model, loader, optimizer, scheduler, device):
    model.train()
    total_loss, all_preds, all_labels = 0, [], []

    for batch in tqdm(loader, desc="Training"):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader)
    acc = accuracy_score(all_labels, all_preds)
    return avg_loss, acc

def eval_epoch(model, loader, device):
    model.eval()
    total_loss, all_preds, all_labels = 0, [], []

    with torch.no_grad():
        for batch in tqdm(loader, desc="Validation"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            total_loss += outputs.loss.item()
            preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader)
    acc = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=le.classes_)
    return avg_loss, acc, report

best_val_acc = 0
for epoch in range(1, EPOCHS + 1):
    print(f"\n{'â'*50}\nEpoch {epoch}/{EPOCHS}")
    train_loss, train_acc = train_epoch(model, train_loader, optimizer, scheduler, device)
    val_loss, val_acc, report = eval_epoch(model, val_loader, device)

    print(f"Train  â  loss: {train_loss:.4f}  acc: {train_acc:.4f}")
    print(f"Val    â  loss: {val_loss:.4f}  acc: {val_acc:.4f}")
    print(report)

    # Save best checkpoint
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_roberta_sentiment.pt')
        print("â Best model saved")
```
